```python
import numpy as np
import pandas as pd
import re
```

```python
fpath='String_Cleaning.xlsx'
df=pd.read_excel(fpath)
df.head(3)
```

```python
df = df.T.reset_index().T
df.columns =['c1','c2']
```

```python
pd.set_option('display.max_colwidth',None)
```

```python
pattern = '([0-9-().]+[0-9])'
df['phone number'] = df['c2'].str.extract(pattern)
```

```python
pattern = '([A-Z]{2})'
df['state'] = df['c2'].str.extract(pattern)
```

```python
pattern = '([A-Z][a-z A-Z]+)'
df['city'] = df['c2'].str.extract(pattern)
```

```python
pattern = '(http[a-z0-9./:-A-Z]+)'
df['url'] = df['c2'].str.extract(pattern)
```

```python
df=pd.read_excel(fpath, header=None)
df.head(3)
```

```python
#  the ? means the preceding pattern is optional (may or may not be present)
phone = '([(]?[0-9]{3}[)|.-]?[0-9]{3}[.-]?[0-9]{4})'
df['phone_number']= df[1].str.extract(phone)
```

```python
# \S is any non-space character
url = '(http[s]?://[\S]+)'
df['url'] = df[1].str.extract(url)
```

```python
city = '([A-Z]{1}[a-z]+[\s]?[A-Za-z]+)'
df['city'] = df[1].str.extract(city)
```

```python
state = '([A-Z][A-Z])'
df['state']= df[1].str.extract(state)
```

```python
df
```

```python
df['phone'] = df['phone_number'].str.strip('(')
```

```python
df['phone'] = df['phone'].str.replace('-', '',regex=False)
```

```python
df['phone'] = df['phone'].str.replace(')', '',regex=False)
```

```python
df['phone'] = df['phone'].str.replace('.', '',regex=False)
```
